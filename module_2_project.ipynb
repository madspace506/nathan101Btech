```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df=pd.read_csv('housing.csv')
```

```python
df.head()
```

```python
sns.scatterplot(x='Landsize', y='Price', data=df, hue='Type' )
plt.show()
```

```python
df.Pruse=df.groupby('YearBuilt')['Price'].mean()
```

```python
sns.lineplot(x=df['YearBuilt'], y=df.Pruse)
plt.show()
```

```python

sns.barplot(x='Regionname', y='Price', data=df, estimator='mean')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.hist(df['Price'])
plt.show()
```

```python

sns.boxplot(x='YearBuilt', y='Regionname', data=df)
plt.xticks(rotation=90)
plt.show()
```
